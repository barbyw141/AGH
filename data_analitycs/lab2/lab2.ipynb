```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from scipy.special import logit, expit
```

```python
data_normal = { 'mean=1 variance =3': np.random.normal(size=1000, scale=3, loc=1),
                'mean=5 variance =3': np.random.normal(size=1000, scale=3, loc=5),
                'mean=10 variance =3': np.random.normal(size=1000, scale=3, loc=10),
                'mean=5 variance =5': np.random.normal(size=1000, scale=5, loc=5),
                'mean=5 variance =8': np.random.normal(size=1000, scale=8, loc=5),
                'mean=5variance =1': np.random.normal(size=1000, scale=1, loc=5)
}
df_norm = pd.DataFrame(data_normal)
melt_norm = df_norm.melt()
```

```python
melt_norm = df_norm.melt()
g = sns.FacetGrid(melt_norm, col='variable')
g.map(plt.hist, 'value')
```

```python
fig, axs = plt.subplots(1, 6)
fig.set_size_inches(18, 5)
axs[0].hist(df_norm.loc[0], cumulative="True")
axs[1].hist(df_norm.loc[1], cumulative="True")
axs[2].hist(df_norm.loc[2], cumulative="True")
axs[3].hist(df_norm.loc[3], cumulative="True")
axs[4].hist(df_norm.loc[4], cumulative="True")
axs[5].hist(df_norm.loc[5], cumulative="True")
```

```python
fig, axs = plt.subplots(1, 6)
fig.set_size_inches(18, 5)
axs[0].hist(df_norm.loc[0], cumulative="True", bins=1)
axs[1].hist(df_norm.loc[1], cumulative="True", bins=2)
axs[2].hist(df_norm.loc[2], cumulative="True", bins=3)
axs[3].hist(df_norm.loc[3], cumulative="True", bins=4)
axs[4].hist(df_norm.loc[4], cumulative="True", bins=5)
axs[5].hist(df_norm.loc[5], cumulative="True", bins=6)
```

```python
df_norm_logit = logit(df_norm)
melt_norm_logit = df_norm_logit.melt()
g = sns.FacetGrid(melt_norm_logit, col='variable')
g.map(plt.hist, 'value')
```

```python
data_poisson = {'lambda = 1': np.random.poisson(size=1000, lam=1),
                'lambda = 5': np.random.poisson(size=1000, lam=5),
                'lambda = 10': np.random.poisson(size=1000, lam=10)
               }
df_poiss = pd.DataFrame(data_poisson)
melt_poiss = df_norm.melt()
g = sns.FacetGrid(melt_poiss, col='variable')
g.map(plt.hist, 'value')
```

```python
data_beta = {'a = 0.1 b = 2': np.random.beta(a=1, b=1, size=1000),
        'a = 1 b= 1': np.random.beta(a=0.3, b=2, size=1000),
        'a = 3 b = 0.5': np.random.beta(a=6, b=3, size=1000)
       }
df_beta = pd.DataFrame(data_beta)
melt_beta = df_beta.melt()
g = sns.FacetGrid(melt_beta, col='variable')
g.map(plt.hist, 'value')
```
